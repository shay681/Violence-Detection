```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [10, 5]
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('./data/CV_DENSE_DIMHAN_5_FOLDS.csv', index_col=[0])
df2 = pd.read_csv('./data/CV_DENSE_DIM2HAN_5_FOLDS.csv', index_col=[0])
```

```python
df = pd.concat([df1,df2])
```

```python
loss_list = []
accuracy_list = []
dims_list = []
dense_dim = df.groupby('dense_dim')[['accaury','loss']].mean()
for index, row in dense_dim.iterrows():
    print(f'For {index} Dense Dim the mean accuary is: {round(row.accaury, 3)} and the mean loss is: {round(row.loss, 3)}.')
    loss_list.append(row.loss)
    accuracy_list.append(row.accaury)
    dims_list.append(index)
```

```python
plt.rcParams['figure.figsize'] = [6, 3]
plt.plot(dims_list,loss_list)
plt.xticks(ticks=dims_list)
plt.xlabel('Dense Layer Units (Num of Dimentions)')
plt.grid()
plt.xlim([8, 256])
plt.ylabel('Loss')
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
learning_data = []
for i in range(df.shape[0]):
    raw_data_string = df.history_df.iloc[i].replace('{', '').replace('}', '').replace('\'', '')
    loss_string_data = raw_data_string.split('[')[1].split(']')[0].replace(' ', '').split(',')
    acc_string_data = raw_data_string.split('[')[2].split(']')[0].replace(' ', '').split(',')
    loss = (np.array(loss_string_data)).astype(float)
    acc = (np.array(acc_string_data)).astype(float)
    fold = df.iloc[i].fold
    dense_dim = df.iloc[i].dense_dim
    for epoch in range(len(acc)):
        epoch_data = {
            "fold": fold,
            "epoch": epoch+1,
            "loss": loss[epoch],
            "acc": acc[epoch],
            "dense_dim": dense_dim
        }
        learning_data.append(epoch_data)
training_df = pd.DataFrame(learning_data)
```

# Dense Layes With 8 Dims

```python
dense_dim = 8
```

```python
epoch_df = training_df[training_df.dense_dim == dense_dim]
x = epoch_df.epoch.unique()
num_of_folds = epoch_df.fold.max()
target = 'acc'
for i in range(num_of_folds):
    k_fold = i + 1
    y = epoch_df[epoch_df['fold'] == k_fold].set_index(['epoch']).drop(['fold'], axis=1)[target]
    plt.plot(x, y, label = f"k = {k_fold}")
plt.xlabel('Epoch')
plt.grid()
plt.ylabel(target.capitalize())
plt.title(target.capitalize() + f' As A Function Of Epochs using Dense laye with {dense_dim} dims')
plt.legend()
plt.show()
```

```python
epoch_df = training_df[training_df.dense_dim == dense_dim]
x = epoch_df.epoch.unique()
num_of_folds = epoch_df.fold.max()
target = 'loss'
for i in range(num_of_folds):
    k_fold = i + 1
    y = epoch_df[epoch_df['fold'] == k_fold].set_index(['epoch']).drop(['fold'], axis=1)[target]
    plt.plot(x, y, label = f"k = {k_fold}")
plt.xlabel('Epoch')
plt.grid()
plt.ylabel(target.capitalize())
plt.title(target.capitalize() + f' As A Function Of Epochs using Dense laye with {dense_dim} dims')
plt.legend()
plt.show()
```

```python
dense_dims = [64, 8, 16, 32, 128, 256]
for dense_dim in dense_dims:
    epoch_df = training_df[training_df.dense_dim == dense_dim]
    target = 'acc'
    x = epoch_df.epoch.unique()
    y = epoch_df.groupby(['epoch'])[target].mean()
    linewidth = 5 if dense_dim == 64  else 2
    color = 'y' if dense_dim == 64  else None
    plt.plot(x, y, linewidth=linewidth, label=f'Dimensions: {dense_dim}', color=color)
    plt.xlabel('Epoch')
    plt.grid()
    plt.xlim([1, 30])
    plt.ylabel('Accuracy')
    plt.legend()
plt.show()
```
