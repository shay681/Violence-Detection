```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
```

# Adam

```python
adam_df = pd.read_csv('./data/HAN_10_ADAM_FOLDS.csv', index_col=[0])
data_df = adam_df
```

```python
print(f'Using ADAM optimaizer with:\nbatch size: {int(data_df.batch_size.mean())}\nepochs: {int(data_df.epochs.mean())}\nlearning rate: {round(data_df.learning_rate.mean(),3)}\nk folds: {data_df.fold.iloc[-1]}\nWe got average loss of:{round(data_df.loss.mean(),3)}, and avarage accaury of {round(data_df.accaury.mean(),3)}% .')
```

```python
learning_data = []
for i in range(data_df.shape[0]):
    raw_data_string = data_df.history_df.iloc[i].replace('{', '').replace('}', '').replace('\'', '')
    loss_string_data = raw_data_string.split('[')[1].split(']')[0].replace(' ', '').split(',')
    acc_string_data = raw_data_string.split('[')[2].split(']')[0].replace(' ', '').split(',')
    loss = (np.array(loss_string_data)).astype(float)
    acc = (np.array(acc_string_data)).astype(float)
    for epoch in range(len(acc)):
        epoch_data = {
            "fold": i+1,
            "epoch": epoch+1,
            "loss": loss[epoch],
            "acc": acc[epoch],
        }
        learning_data.append(epoch_data)
training_df = pd.DataFrame(learning_data)
```

```python
x = training_df.epoch.unique()
num_of_folds = training_df.fold.max()
target = 'loss'
for i in range(num_of_folds):
    k_fold = i + 1
    y = training_df[training_df['fold'] == k_fold].set_index(['epoch']).drop(['fold'], axis=1)[target]
    plt.plot(x, y, label = f"k = {k_fold}")
plt.xlabel('Epoch')
plt.grid()
plt.ylabel(target.capitalize())
plt.title(target.capitalize() + ' As A Function Of Epochs')
plt.legend()
```

```python
target = 'loss'
x = training_df.epoch.unique()
y = training_df.groupby(['epoch'])[target].mean()
plt.plot(x, y, linewidth=4)
plt.xlabel('Epoch')
plt.grid()
plt.ylabel(target.capitalize())
plt.title('Average ' + target.capitalize() + ' As A Function Of Epochs')
```

```python
x = training_df.epoch.unique()
num_of_folds = training_df.fold.max()
target = 'acc'
for i in range(num_of_folds):
    k_fold = i + 1
    y = training_df[training_df['fold'] == k_fold].set_index(['epoch']).drop(['fold'], axis=1)[target]
    plt.plot(x, y, label = f"k = {k_fold}")
plt.xlabel('Epoch')
plt.grid()
plt.ylabel(target.capitalize())
plt.title(target.capitalize() + ' As A Function Of Epochs')
plt.legend()
```

```python
target = 'acc'
x = training_df.epoch.unique()
y = training_df.groupby(['epoch'])[target].mean()
plt.plot(x, y, linewidth=4)
plt.xlabel('Epoch')
plt.grid()
plt.ylabel(target.capitalize())
plt.title('Average ' + target.capitalize() + ' As A Function Of Epochs')
```
